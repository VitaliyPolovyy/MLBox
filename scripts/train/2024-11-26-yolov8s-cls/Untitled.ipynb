```python
from pathlib import Path
import os
import torch
import numpy as np
from ultralytics import YOLO
```

```python
model = YOLO( "yolov8s-cls.pt")


input_folder = Path("peanuts_class")
yaml_file = Path("/workspace/peanuts_class/dataset.yaml")

print(str (input_folder.resolve()),)


model.train(
    data=str (input_folder.resolve()),
    epochs=100,
    imgsz=128,
    batch=16,
    device=0 if torch.cuda.is_available() else "cpu",
    workers=8,
    task="classify",
    
    fliplr=0.5,    # Horizontal flip (50% probability)
    flipud=0.2,    # Vertical flip (20% probability)
    degrees=15,    # Rotation ±15 degrees
    scale=0.1,     # Scaling ±10%
    shear=5,       # Slight shearing
)
```
